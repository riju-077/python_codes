*Exercise 3 — Real API-style response (deeper nesting)*

Questions:
  1. Get Agnik's email
  2. Get the amount of Agnik's second order (3200)
  3. Get Rahul's phone number
  4. Loop through all users and print their name and email
  5. Loop through Agnik's orders and print each order_id and amount


```python
response_2  =  {
      "page": 1,
      "total_pages": 5,
      "users": [
          {
              "id": 1,
              "name": "Agnik",
              "contacts": {"email": "agnik@test.com", "phone": "9876543210"},
              "orders": [
                  {"order_id": "A101", "amount": 1500},
                  {"order_id": "A102", "amount": 3200}
              ]
          },
          {
              "id": 2,
              "name": "Rahul",
              "contacts": {"email": "rahul@test.com", "phone": "9123456780"},
              "orders": [
                  {"order_id": "B201", "amount": 800}
              ]
          }
      ]
  }
```

```python
print(response_2["users"][0]["contacts"]["email"])
print(response_2["users"][0]["orders"][1]["amount"])
print(response_2["users"][1]["contacts"]["phone"])
```

for item in response_2["users"]:
  print(f"{{order_id : {item["orders"]["order_id"]} , amount: {item["orders"]["amount"]} }}")

```python
for order in response_2["users"][0]["orders"]:
    print(f"{{order_id: {order['order_id']}, amount: {order['amount']} }}")
```

```python
result = [{"order_id": order["order_id"], "amount": order["amount"]} for order in response_2["users"][0]["orders"]]
print(result)
```

```python
result = [{"name": item["name"], "email": item["contacts"]["email"]} for item in response_2["users"]]
print(result)
```
